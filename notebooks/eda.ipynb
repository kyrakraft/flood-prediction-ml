```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

#Set paths for importing data
project_root = os.path.abspath(os.path.join(os.getcwd(), "..")) #set root to main project dir
data_path = os.path.join(project_root, 'data') #path to data folder

#Load data
train = pd.read_csv(os.path.join(data_path, 'train.csv'))
test = pd.read_csv(os.path.join(data_path, 'test.csv'))

train = train.drop('id', axis=1) 

#Basic info

print("Train shape:", train.shape)
print("Test shape:", test.shape)
print("\nMissing values:\n", train.isnull().sum())

train.head()
```

```python
train.describe().T
```

```python
#EDA

corr_matrix = train.corr(numeric_only=True) #compute correlation matrix

target_corr = corr_matrix['FloodProbability'].sort_values(ascending=False)
print(target_corr)

#Plot heatmap
plt.figure(figsize=(12,10))
sns.heatmap(corr_matrix, cmap='coolwarm', center=0, annot=False)
plt.title('Feature Correlation Matrix')
plt.show()
```

```python
sns.clustermap(corr_matrix, cmap='coolwarm', figsize=(12,12))
```

```python
train.hist(figsize=(16, 12), bins=30, edgecolor='black')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))
sns.histplot(train['FloodProbability'], kde=True, bins=50)
plt.title("Distribution of FloodProbability")
plt.show()
```

```python
import xgboost as xgb
X = train.drop(columns='FloodProbability')
y = train['FloodProbability']

model = xgb.XGBRegressor(n_estimators=50, max_depth=4, random_state=42)
model.fit(X, y)

xgb.plot_importance(model, importance_type='gain', max_num_features=30, title='Feature Importance (by Gain)')
plt.tight_layout()
plt.show()
```

```python
import xgboost as xgb
X_temp = train.drop(columns='FloodProbability')
X = X_temp
y = train['FloodProbability']

X['sum_of_features'] = X_temp.sum(axis=1)

X['row_mean'] = X_temp.mean(axis=1)
X['row_median'] = X_temp.median(axis=1)
X['row_std'] = X_temp.std(axis=1)
X['row_q25'] = X_temp.quantile(0.25, axis=1)
X['row_q75'] = X_temp.quantile(0.75, axis=1) #don't need bc all have val of 6.00?
X['row_max'] = X_temp.max(axis=1)
X['row_min'] = X_temp.min(axis=1)

model = xgb.XGBRegressor(n_estimators=50, max_depth=4, random_state=42)
model.fit(X, y)

xgb.plot_importance(model, importance_type='gain', max_num_features=50, title='Feature Importance (by Gain)')
plt.tight_layout()
plt.show()
```

```python
import xgboost as xgb
X_temp = train.drop(columns='FloodProbability')
X = pd.DataFrame(index=train.index)
y = train['FloodProbability']

X['sum_of_features'] = X_temp.sum(axis=1)

X['row_mean'] = X_temp.mean(axis=1)
X['row_median'] = X_temp.median(axis=1)
X['row_std'] = X_temp.std(axis=1)
X['row_q25'] = X_temp.quantile(0.25, axis=1)
X['row_q75'] = X_temp.quantile(0.75, axis=1) #don't need bc all have val of 6.00?
X['row_max'] = X_temp.max(axis=1)
X['row_min'] = X_temp.min(axis=1)

model = xgb.XGBRegressor(n_estimators=50, max_depth=4, random_state=42)
model.fit(X, y)

xgb.plot_importance(model, importance_type='gain', max_num_features=30, title='Feature Importance (by Gain)')
plt.tight_layout()
plt.show()
```
